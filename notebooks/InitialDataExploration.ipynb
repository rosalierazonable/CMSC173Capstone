# Preliminary Data Analysis on VBDs

```python
# adding a separate column for date only
df['Date'] = df['Date_Proc'].apply(lambda x: x[:10])
```

```python
# count the number of VBDs recorded per day
counts_per_day = df.groupby('Date')['id'].agg('count')

# plotting it
plt.figure(figsize=(10,10))
plt.bar(counts_per_day.keys(), counts_per_day)
plt.xticks(range(0,31), [date[5:] for date in counts_per_day.keys()], rotation=90)
plt.xlabel('Date')
plt.ylabel('VBDs')
plt.title('January VBDs per day')
plt.plot()
```

```python
# print the date and time the sensor has recorded a value
# print the number of VBDs recorded per date-time
date = ''
counts = df.groupby('Date_Proc')['id'].agg('count')
for key in counts.keys():
    if key[:10] != date:
        date = key[:10]
        print(date)
    print('\t{} - {}'.format(key[10:],counts[key]))
```

```python
philippines = geopandas.read_file('../../samplefiles/PHL_adm0.shp')
philippines = philippines['geometry']
```

```python
philippines.plot(figsize=(10,10))
plt.show()
```

```python
# for scaling purposes
EEZ = geopandas.read_file('../../samplefiles/eez.shp')
```

```python
records = geopandas.read_file('../../samplefiles/2021January.shp')
records = records[records['EEZ'] == 'Philippines Exclusive Economic Zone']
```

```python
fig,ax = plt.subplots(figsize=(15,15))
EEZ.plot(ax=ax, alpha=0.1, color='green')
philippines.plot(ax=ax, alpha=0.4, color='grey')
records.plot(ax=ax, color='red')
plt.title('All VBDs for January 2021')
```

```python
# optional, do not run if not needed
# create pictures for animation
# https://towardsdatascience.com/how-to-make-a-gif-map-using-python-geopandas-and-matplotlib-cd8827cefbc8

for i in ['0'+str(i) if i < 10 else str(i) for i in range(1,32)]:
    subrecord = records[records['Date_Proc'].str.contains('2021-01-'+i)]
    fig,ax = plt.subplots(figsize=(15,15))
    EEZ.plot(ax=ax, alpha=0.1, color='green', zorder=1)
    philippines.plot(ax=ax, alpha=0.4, color='grey', zorder=2)
    subrecord.plot(ax=ax, color='red', zorder=3)
    plt.title('All VBDs for 2021-01-'+i)
    filepath=os.path.join('../../Visuals','2021-01-'+i+'.jpg')
    fig.savefig(filepath, dpi=100)
```
