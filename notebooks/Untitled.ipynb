```python
import cv2 as cv
from PIL import Image
from matplotlib import pyplot as plt
from scipy import ndimage
```

```python
def plot(imagesrc):
    plt.figure(figsize=(10,10))
    plt.imshow(imagesrc)
    plt.plot()
```

```python
img = cv.imread('../Satellite/satimg.tiff')
plot(img)
```

```python
img_gray = cv.imread('../Satellite/satimg.tiff', cv.IMREAD_GRAYSCALE)
plot(img_gray)
```

```python
img_gray2 = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plot(img_gray2)
```

```python
from scipy.signal import wiener

img_wiener = wiener(img_gray*1000000)
```

```python
f, (plot1, plot2) = plt.subplots(1,2, figsize=(25,25))
plot1.imshow(img_wiener)
plot2.imshow(img_gray)
plt.show()
```
